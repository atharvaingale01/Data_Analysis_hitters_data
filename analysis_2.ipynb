# Solutions

## Hitters

### Find of people who have batted for more than 400 matches

```python
df2[df2.CAtBat >400][['Name','AtBat']]
```

### Find out the name of the player who has hit max home runs

```python
df2[df2.CHmRun == df2.CHmRun.max()][["Name","CHmRun"]]
```

### Find out league wise total number of players

```python
df2.groupby(by = "League").Name.count()
```

### Find out division wise total amount of salary invested

```python
df2.groupby(by = 'Division').Salary.sum()
```

### Find out five most senior players

```python
df2[["Name","Years"]].sort_values(by = "Years",ascending=False).head()
```

## Cars

### Find out car having highest Mileage in city

```python
df[df["MPG.city"] == df["MPG.city"].max()][["Make","MPG.city"]]
```

### Find out Models of cars which are having Price between 17 and 25

```python
df[(df.Price > 17) & (df.Price < 25)][["Model",'Price']]
```

### Find out top 7 most fuel economic cars for highway

```python
df[["Make","MPG.highway"]].sort_values(by = "MPG.highway",ascending=False).head(7)
```

### Find out Models of cars having no AirBags and which are of compact type whose price is below 20.

```python
df["AirBags"].unique()
```

```python
df[((df.AirBags == "None") | (df.AirBags.isna())) & (df.Type == 'Compact') & (df.Price <20)]
```

### Find out top 7 most fuel economic cars for city, show results in sorted manner based on Horsepower

```python
df5 = df.sort_values(by = "MPG.city",ascending=False).head(7)
```

```python
df5.sort_values(by = "Horsepower")
```

### Find out Cars having mileage of more than 15 miles in city

```python
df[df["MPG.city"] > 15]
```

### Find out Cars with US based origin

```python
df[df.Origin == "USA"]
```

### Find out Cars with no airbags and which are compact in type, from the output retrieve only Model, Mileage in city, Type, Airbags

```python
df[((df.AirBags == "None") | (df.AirBags.isna())) & (df.Type == 'Compact')][["Model", "MPG.city", "Type", "AirBags"]]
```

### Find out cars with Mileage in city more than 15 which are compact in type

```python
df[(df["MPG.city"] > 15) & (df.Type == 'Compact')]
```
