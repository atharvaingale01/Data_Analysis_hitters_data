Create a new column called as Car_Cat
1. If horsepower is less than 100....value "LS"
2. If horsepower is btn(101-200)....value "MS"
3. If horsepower is above 200....value "HS"

```python
Car_Cat = []

for i in range(93):
    if A.Horsepower[i] < 100:
        Car_Cat.append("LS")
    elif (A.Horsepower[i] >= 101) & (A.Horsepower[i] <= 200):
        Car_Cat.append("MS")
    else:
        Car_Cat.append("HS")

B = pd.DataFrame(Car_Cat,columns =['Car_Cat'])
Result = pd.concat([A,B], axis=1)
Result
```
